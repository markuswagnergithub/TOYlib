```python
file="KinderSurprise2019data.csv"
```

```python
import csv
```

Read file:

```python
with open(file, mode='r') as f:
    reader = csv.reader(f)
    next(reader) #skip header
    itemvaluesCSV = {(int(rows[5]),int(rows[2])) : float(rows[3]) for rows in reader}
```

```python
from gurobipy import *
```

Use Gurobi's multidict function to generate keys and a dictionary:

```python
items, itemvalues = multidict(itemvaluesCSV)
```

```python
items
```

```python
itemvalues
```

```python
m = Model('toys')
```

Add decision variables:

```python
vars = m.addVars(items,
                 vtype=GRB.BINARY,
                 name="x")
```

```python
vars
```

Add constraints: within each set i, only one item can be allocated 

```python
m.addConstrs((vars.sum(i,'*') == 1 for i in range(1,188,1)), name='useonce')
```

Define the objective:

```python
obj = vars.prod(itemvalues)
```

```python
obj
```

```python
m.setObjective(obj,GRB.MAXIMIZE)
```

```python
m.optimize()
```

Write models to file for later inspection:

```python
m.write('toys.lp')
m.write('toys.mps')
```

Output the values of the decision variables:

```python
for v in m.getVars():
  #if v.x != 0:
    print('%s %g' % (v.varName, v.X))
```

Output the objective value:

```python
obj=m.getObjective()
print('Obj: %g' % obj.getValue())
```
